#### 리스트 자료형

일반적인 !!!!! 배열 (Array) , 리스트 (List) 의 차이

##### 배열
* 배열은 동일한 자료형의 원소들을 "연속적인 메모리 공간에" 저장하는 자료 구조이며, 배열의 크기는 고정되어 있고, 선언 시에 크기를 지정해야함

##### 리스트
* 리스트는 원소들을 연결하여 저장하는 자료 구조로서, 원소의 개수가 가변적이며 삽입과 삭제가 자유로움

##### 차이점
* 메모리 할당 : 배열은 연속적, 리스트는 비연속적
* 크기 : 배열은 고정, 리스트는 가변적
* 접근 방법 : 배열은 인덱스, 리스트는 순차적
* 삽입과 삭제 : 배열은 번거로움, 리스트는 빠름

##### (참고)파이썬에서 list 는 연결 리스트가 아니기 때문에 위와 같은 방식으로 적용이 되지 않음.

````python
dic = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
````

````python
## 딕셔너리를 만들 때 주의 사항
### Key 값은 고유한 값이므로, 중복되는 Key 값을 설정해놓으면 하나를 제외한 나머지 것들이 모두 무시됨

a= {1:'a', 1:'b'}
print(a[1])
````

````python

marks = [90, 25, 67, 45, 80]   # 학생들의 시험 점수 리스트

number = 0   # 학생에게 붙여 줄 번호
for mark in marks:   # 90, 25, 67, 45, 80을 순서대로 mark에 대입
    number = number +1
    if mark >= 60:
        print("%d번 학생은 합격입니다." % number)
    else:
        print("%d번 학생은 불합격입니다." % number)
````

````python
## for 문과 continue 문
# marks2.py
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
    number = number +1
    if mark < 60:
        continue
    print("%d번 학생 축하합니다. 합격입니다. " % number)
````

````python
## for 문과 같이 자주 쓰이는 range 함수
### 리스트를 자동으로 만들어줌
### range(시작 숫자, 끝 숫자)
a = range(10)
print(a)
b = range(1,11)
print(b)
````

````python
# 1~10 까지 더하는 것을 구하려면?
add = 0
for i in range(1, 11):
   add = add + i

print(add)
````

````python
# len 함수를 같이 써보기
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print("%d번 학생 축하합니다. 합격입니다." % (number+1))
````

````python
### ** list comprehension 알아두기
## 리스트를 더 직관적으로 활용하는 방법

## 예를들어 a=[1,2,3,4] 이고 각 원소에 3 씩 곱해주고 싶다면 기존에는
a=[1,2,3,4]
result = []
for num in a:
  result.append(num*3)
print(result)
````

````python
# list comprehension 을 활용한다면
a = [1,2,3,4]
result= [num*3 for num in a]
print(result)
````

````python
# 조금 더 복잡하게 하자
# [표현식 for 항목 in 반복_가능_객체 if 조건문]
result = [x*y for x in range(2,10)
              for y in range(1,10)]
print(result)
````

#### 함수

함수는 특정 작업을 수행하는 "코드의 집합"입니다. 클라우드에서는 AWS Lambda나 Google Cloud Functions 같은 Serverless 서비스의 최소 단위가 됩니다.

````python
# [실습] 인스턴스 상태를 확인하는 함수 정의
def check_status(instance_id, status="running"):
    """인스턴스의 ID와 상태를 출력하는 함수"""
    print(f"[LOG] 인스턴스 {instance_id}의 현재 상태는 {status}입니다.")
    return f"Status of {instance_id}: {status}"

# 함수 호출
res = check_status("i-0abc123")
check_status("i-0xyz789", "stopped")
print("print(res) 결과 --> ", res)


# 가변 매개변수 (*args): 여러 대의 서버 ID를 한 번에 받을 때 유용
def terminate_instances(*ids):
    for instance_id in ids:
        print(f"Danger: {instance_id}를 종료합니다.")

terminate_instances("srv-01", "srv-02", "srv-03")
````

#### 클래스

클래스는 **"붕어빵 틀"**입니다. 클라우드에서는 **'Virtual Machine'**이라는 설계도(Class)를 가지고 수많은 '인스턴스'(Object)를 찍어내는 과정과 같습니다.

````python
class CloudInstance:
    # 생성자: 인스턴스가 만들어질 때 초기화 (메모리 할당)
    def __init__(self, name, cpu, ram):
        self.name = name
        self.cpu = cpu
        self.ram = ram
        self.is_running = False

    # 메서드: 인스턴스가 할 수 있는 행동
    def start(self):
        self.is_running = True
        print(f"서버 {self.name}이 시작되었습니다. (CPU: {self.cpu} Core)")

    def stop(self):
        self.is_running = False
        print(f"서버 {self.name}이 중지되었습니다.")

# 객체 생성 (인스턴스화)
web_server = CloudInstance("Web-Srv-01", 2, 4096)
db_server = CloudInstance("DB-Srv-01", 4, 8192)

web_server.start()
print(f"DB 서버 사양: {db_server.ram}MB")
````
